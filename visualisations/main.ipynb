# Visualisations for the results

```python
# imports

import seaborn as sns
import json
import pandas as pd
import matplotlib.pyplot as plt 

# consts

SYSTEM_ELO = "system elo"
ELO_HISTORY = "elo history"
GAME_EVAL_HISTORY = "game evaluation history"
GAME_STAT_HISTORY = "game status history"
TIME_PER_MOVE = "time per move"
MAX_MOVES = 30
GAMES = [f"{i}" for i in range(10)]

# reading data

with open(r"C:\Users\Gjorgji\Projects\chess-algorithm-comparison\genetic.json") as file:
    res_genetic = json.load(file)
with open(r"C:\Users\Gjorgji\Projects\chess-algorithm-comparison\genetic2.json") as file:
    res_genetic2 = json.load(file)
with open(r"C:\Users\Gjorgji\Projects\chess-algorithm-comparison\minimax.json") as file:
    res_minimax = json.load(file)
with open(r"C:\Users\Gjorgji\Projects\chess-algorithm-comparison\monte_carlo.json") as file:
    res_monte_carlo = json.load(file)
with open(r"C:\Users\Gjorgji\Projects\chess-algorithm-comparison\monte_carlo_30its.json") as file:
    res_monte_carlo_30its = json.load(file)

# setups

sns.set_style("darkgrid")
```

### Minimax with alpha beta pruning

```python
avg_times = [res_minimax[TIME_PER_MOVE][key] for key in GAMES]
temp_times_per_move = [[a[i] for a in avg_times if len(a) > i] for i in range(MAX_MOVES)]
avg_time_per_move_minimax = [sum(i) / len(i) for i in temp_times_per_move]

plt.xlabel("Move number", size=13)
plt.ylabel("Avg time per move (s)", size=13)
plt.title("Average time per move for minimax at depth 3", size=15)
sns.lineplot(y=avg_time_per_move_minimax, x=range(1, 31))
```

### Monte Carlo with 30 iterations

```python
avg_times = [res_monte_carlo_30its[TIME_PER_MOVE][key] for key in GAMES]
temp_times_per_move = [[a[i] for a in avg_times if len(a) > i] for i in range(MAX_MOVES)]
avg_time_per_move_monte_carlo_30_its = [sum(i) / len(i) for i in temp_times_per_move if len(i) > 0]

plt.xlabel("Move number", size=13)
plt.ylabel("Avg time per move (s)", size=13)
plt.title("Average time per move for Monte Carlo at 30 iterations", size=13)
sns.lineplot(y=avg_time_per_move_monte_carlo_30_its, x=range(1, 28))
```

### Monte carlo with 10 iterations

```python
avg_times = [res_monte_carlo[TIME_PER_MOVE][key] for key in GAMES]
temp_times_per_move = [[a[i] for a in avg_times if len(a) > i] for i in range(MAX_MOVES)]
avg_time_per_move_monte_carlo_10_its = [sum(i) / len(i) for i in temp_times_per_move if len(i) > 0]

plt.xlabel("Move number", size=13)
plt.ylabel("Avg time per move (s)", size=13)
plt.title("Average time per move for Monte Carlo at 10 iterations", size=13)
sns.lineplot(y=avg_time_per_move_monte_carlo_10_its, x=range(1, 31))
```

### Combined results

```python
plt.xlabel("Move number", size=13)
plt.ylabel("Avg time per move (s)", size=13)
plt.title("Average time per move for Monte Carlo and Minimax", size=13)
sns.lineplot(y=avg_time_per_move_monte_carlo_10_its, x=range(1, 31))
sns.lineplot(y=avg_time_per_move_monte_carlo_30_its, x=range(1, 28))
sns.lineplot(y=avg_time_per_move_minimax, x=range(1, 31))
plt.legend(labels=["Monte Carlo (10 its.)", "Monte Carlo (30 its.)", "Minimax"])
```
